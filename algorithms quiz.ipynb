```python
'''Quiz: integer array that outputs a given pair that sums up to a given value'''
def array(a,k):
    if len(a)<2:
        print('too short')
        
    seen=set()
    output=set()
    
    for n in a:
        target=k-n
        
        if target not in seen:
            seen.add(n)
        else:
            output.add((min(n,target),max(n,target)))
    print('\n'.join(map(str,list(output))))
```

```python
array([1,9,5,5,6,4],10)
```

```python
'''common elements(array) in 2 arguments'''
def common(a,b):
    p1=0
    p2=0
    
    results=[]
    
    while p1<len(a) and p2<len(b):
        
        if a[p1]==b[p2]:
            results.append(a[p1])
            p1+=1
            p2+=2
        
        elif b[p2]<a[p1]:
            p2+=1
            
        else:
            p1+=1
            
    return results
```

```python
common([1,2,4],[0,2,5])
```

```python
'''Big O notation -graph representation'''

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('bmh')
n=np.linspace(1,10)

labels=['constant','linear','logarithm','log linear','cubic','quandratic','exponential']

big_O=[np.ones(n.shape),n,np.log(n),n*np.log(n),n**2,n**3,2**n]

plt.figure(figsize=(12,10))

plt.ylim(0,50)

for i in range(len(big_O)):
    plt.plot(n,big_O[i],label=labels[i])
    
plt.legend(loc=0)

plt.xlabel('n')
plt.ylabel('Relative Runtime')
```

```python
'''dynamic array'''
def anagram(s1,s2):
    s1=s1.replace(' ','').lower()
    s2=s2.replace(' ','').lower()
    
    count={}
    
    if len(s1) != len(s2):
        return False
    
    for i in s1:
        if i in count:
            count[i]+=1
        else:
            count[i]=1
            
    for j in s2:
        if j in count:
            count[j]-=1
        else:
            count[j]=1
            
    for k in count:
        if count[k]!=0:
            return False
    return True
    
    
```

```python
anagram('meat','TE AM')
```

```python
'''frequent count'''
'''given an list find the most frequent count'''
def count(list):
    count={}
    max_count=0
    max_item=None
    
    for i in list:
        if i in count:
            count[i]+=1
        else:
            count[i]=1
            
        if count[i]>max_count:
            count[i]=max_count
            max_item=i
    
    return max_item
```

```python
count([1,2,3,5,3])
```

```python
'''largest sum'''
'''find the largest sum in an array'''

def maximum(arr):
    if len(arr)==0:
        print('too small')
        
    current_sum=max_sum=arr[0]
    
    for i in arr[1:]:
        current_sum=max(current_sum+i,i)
        max_sum=max(max_sum,current_sum)
        
    return max_sum
```

```python
maximum([1,2,3])
```

```python
'''unique characters'''
def unique(s):
    s=s.replace(' ','').lower()
    
    count={}
    all_unique=[]
    
    for i in s:
        if i in count:
            count[i]+=1
        else:
            count[i]=1
            
    y=sorted(count.items(),key= lambda x:x[1])
    
    for i in y:
        if i[1]==y[0][1]:
            all_unique.append(i)
        
    return all_unique
    
```

```python
'''reverse string'''
def reverse(s):
    
    i=0
    l=len(s)
    space=[' ']
    words=[]
    
    while i<l:
        if s[i] not in space:
            word_start=i
            
            while i<l and s[i] not in space:
                i+=1
                words.append(s[word_start:i])
                
    return ''.join(reversed(s)) 
```
